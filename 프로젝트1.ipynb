```python
import numpy as np
import csv
import re
import matplotlib.pyplot as plt
```

```python
movie_path="/IMDB top 1000.csv"
```

```python
data_list=[]
with open(movie_path, "r", encoding="utf-8") as f:
  reader=csv.reader(f)
  header=next(reader)
  for row in reader:
    data_list.append(row)


  data_array=np.array(data_list)

  title_idx=header.index("Title")
  genre_idx=header.index("Genre")
  rate_idx=header.index("Rate")

  selected_data=data_array[:,[title_idx, genre_idx, rate_idx]]
  filtered_data=selected_data[~np.any(selected_data=='', axis=1)]

processed = []

for row in data_array:
  title = row[title_idx]
  genre = row[genre_idx]
  rate = row[rate_idx]

  Released_Year = title[-5:-1]
  processed.append([title, genre, rate, Released_Year])

  final_array=np.array(processed)
  print(final_array[:3])
  print("데이터 shape:", final_array.shape)

  years = final_array[:, 3]
  rates = final_array[:, 2].astype(float)
  unique_years = np.unique(years)
  yearly_avg = []
  for year in unique_years:
    indices = np.where(years == year)
    year_ratings = rates[indices]
    avg_rating = np.mean(year_ratings)
    yearly_avg.append([year, avg_rating])
    yearly_avg_array = np.array(yearly_avg, dtype=object)
    sorted_result = yearly_avg_array[np.argsort(yearly_avg_array[:, 0])]

  for year, avg in sorted_result:
    print(f"{year}: {round(float(avg), 2)}")

```
